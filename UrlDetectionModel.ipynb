```python
import pandas as pd
import numpy as np
```

```python
phising_site_url = pd.read_csv('./phishing_site_urls.csv')
phising_site_url = phising_site_url.dropna(axis=0)
phising_site_url
```

```python
good_site = phising_site_url[phising_site_url['Label'] == 'good']
bad_site = phising_site_url[phising_site_url['Label'] == 'bad']
print(good_site.Label.value_counts())
print(bad_site.Label.value_counts())
```

```python
malicious_phish1 = pd.read_csv('./malicious_phish1.csv')
malicious_phish1
```

```python
malicious_phish1 = malicious_phish1.dropna(axis = 0)
malicious_phish1.type.value_counts()
```

```python
benign_site = malicious_phish1[malicious_phish1['type'] == 'benign']
nonbenign_site = malicious_phish1[malicious_phish1['type'] != 'benign']

benign_site['type'] = benign_site['type'].replace(['benign'], 'good')
nonbenign_site['type'] = nonbenign_site['type'].replace(['defacement','phishing','malware'], 'bad')

print(benign_site.type.value_counts())
print(nonbenign_site.type.value_counts())
```

```python
urlhaus = pd.read_csv('./urlhaus.csv')
urlhaus
```

```python
urlhaus = urlhaus.dropna(axis = 0)
urlhaus.threat.value_counts()
```

```python
good_site = good_site.rename(columns={'Label':'type'})
good_site = good_site.rename(columns={'URL':'url'})
bad_site = bad_site.rename(columns={'Label':'type'})
bad_site = bad_site.rename(columns={'URL':'url'})
bad_site
```

```python
urlhaus['type'] = 'bad'
urlhaus = urlhaus.drop(['# id', 'dateadded', 'urlhaus_link' , 'url_status', 'last_online', 'threat', 'tags', 'reporter'], axis=1)
urlhaus
```

```python
#악성 url 데이터셋
test_data = pd.concat([bad_site, urlhaus])
test_data = pd.concat([test_data, nonbenign_site])
test_data
```

```python
#정상 url
data = pd.concat([benign_site,good_site])
data
```

```python
#정상 데이터 url 길이 측정
midLen = 0;
count = 0
for index, row in data.iterrows():
    midLen += len(row['url'])
    if len(row['url']) < 20:
        count += 1

print(count)
print(len(data))
print(midLen)
print(midLen/len(data))
```

```python
#악성 데이터 url 길이 측정
testMidLen = 0;
testCount = 0;
for index, row in test_data.iterrows():
    testMidLen += len(row['url'])
    if len(row['url']) < 20:
        testCount += 1
        
    
print(testCount)
print(len(test_data))
print(testMidLen)
print(testMidLen/len(test_data))
```

```python
from urllib.parse import urlparse
import re


new_match = []
for index, row in data.iterrows():
    url = row['url']
    parsed_url = urlparse(url)
    host = parsed_url.netloc
    if re.match(r'^\d+\.\d+\.\d+\.\d+:\d+$', host) :
        new_match.append(True)
    else :
        new_match.append(False)

data['IPT/F'] = new_match
data
```

```python
import socket
from urllib.parse import urlparse

new_test_data_dns = []

for index, row in test_data.iterrows():
    # URL 정의
    url = row['url']
    # urlparse를 사용하여 호스트 추출
    
    parsed_url = urlparse(url)
    host = parsed_url.netloc

    # DNS 질의 수행
    try:
        ip_address = socket.gethostbyname(host)
        new_test_data_dns.append(True)
#         print(f"IP Address: {ip_address}")
    except socket.gaierror:
        new_test_data_dns.append(False)
#         print("DNS lookup failed")

        
new_test_data_dns
# test_data['DNST/F'] = new_test_data_dns
# test_data
```

```python
import dns.resolver

new_test_data_ttl = []

test_ttl = test_data[test_data['lenUrl'] < 63]

for index, row in test_ttl.iterrows():
    
    host = row['url']

    resolver = dns.resolver.Resolver()
    resolver.timeout = 1
    resolver.lifetime = 10

    try:
        query_result = resolver.resolve(host)
        for rdata in query_result:
            new_test_data_ttl.append(True)
    except dns.resolver.NoAnswer:
        new_test_data_ttl.append(False)
    except dns.resolver.NXDOMAIN:
        new_test_data_ttl.append(False)
    except dns.resolver.NoNameservers:
        new_test_data_ttl.append(False)
    
    
test_ttl['TTLT/F'] = new_test_data_ttl
test_ttl
```

```python
import dns.resolver

test_ns = test_data[test_data['lenUrl'] < 63]

for index, row in test_ns.iterrows():

    domain = row['url']

    resolver = dns.resolver.Resolver()
    resolver.timeout = 1
    resolver.lifetime = 10

    try:
        query_result = resolver.query(domain, "NS")
        nameservers = [str(ns) for ns in query_result]
        print(f"Nameservers for {domain}:")
        for nameserver in nameservers:
            print(nameserver)
    except dns.resolver.NoAnswer:
        print("No NS record found")
    except dns.resolver.NXDOMAIN:
        print("Domain does not exist")
    except dns.resolver.NoNameservers:
        print("NonameServer")

```

```python
import dns.name

test_ns = test_data[test_data['lenUrl'] < 63]

for index, row in test_ns.iterrows():

    domain = row['url']


    try:
        # 도메인 이름을 파싱하여 DNS 레이블을 검사합니다.
        parsed_domain = dns.name.from_text(domain)
        labels = parsed_domain.labels

        # 빈 레이블을 제거하여 수정된 도메인 이름을 생성합니다.
        modified_domain = ".".join(label.decode() for label in labels if label)

        print(f"Modified domain: {modified_domain}")
    except dns.name.EmptyLabel:
        print("Invalid domain: EmptyLabel")
```
